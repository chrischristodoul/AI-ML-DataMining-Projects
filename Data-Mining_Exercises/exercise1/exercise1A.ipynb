Ανάλυση: Reservoir Sampling

Στο παρόν κελί κειμένου, γίνεται η ανάλυση του τροποποιημένου αλγορίθμου του Reservoir Sampling σε απλά ελληνικά.

1.Ο αλγόριθμος κρατά μια δεξαμενή (reservoir) μεγέθους Κ και έναν μετρητή n που δείχνει πόσα αντικείμενα έχουν περάσει μέχρι στιγμής από το ρεύμα.

2.Ο αλγόριθμος ξεκινά να διατρέχει τα στοιχεία του ρεύματος ένα προς ένα και αποθηκεύει τα Κ πρώτα αντικείμενα που θα συναντήσει. Μόλις γεμίσει η «δεξαμενή», ο μετρητής θα αρχικοποιηθεί με τιμή Κ. Αν το ρεύμα σταματήσει τώρα, ο αλγόριθμος θα επέστρεφε τα Κ αντικείμενα με πιθανότητα το καθένα να είναι 1, δηλαδή P=(K/n)=(K/K)=1

3.Αν όμως το ρεύμα δε σταματήσει, τότε ο αλγόριθμος συνεχίζει στα επόμενα στοιχεία–αντικείμενα στις θέσεις Κ+1, Κ+2, … και ενημερώνει την τιμή του n με τα όσα στοιχεία έχουν διαβαστεί από το ρεύμα. Με το που βρει το επόμενο αντικείμενο, ο αλγόριθμος το επιλέγει με πιθανότητα K/n όπου n είναι η τρέχουσα τιμή του μετρητή — για παράδειγμα, για το αντικείμενο στη θέση Κ+1 έχουμε πιθανότητα K/(K+1).Το απορρίπτει με πιθανότητα 1−K/n, οπότε αγνοείται και συνεχίζουμε με το επόμενο αντικείμενο του ρεύματος.

4.Αν επιλεγεί, τότε θα πρέπει να τοποθετηθεί μέσα στη «δεξαμενή», αντικαθιστώντας ένα από τα υπάρχοντα στοιχεία, το οποίο επιλέγεται ομοιόμορφα τυχαία μεταξύ των Κ θέσεων (πιθανότητα 1/Κ για καθεμία).

5.Μετά την ανταλλαγή των δύο αντικειμένων, την απόρριψη του ενός και την αποθήκευση του άλλου στη «δεξαμενή», ο αλγόριθμος εκτελεί κάθε φορά επαναληπτικά τα βήματα 3 και 4 μέχρι να εξαντληθεί το ρεύμα. Όταν σταματήσει, τότε ο τροποποιημένος αλγόριθμος θα επιστρέψει τα Κ αντικείμενα που βρίσκονται στη «δεξαμενή» εκείνη τη στιγμή, με το καθένα να έχει πιθανότητα K/N.

Έτσι αποδεικνύουμε ότι κάθε αντικείμενο που βρίσκεται εντός της «δεξαμενής» έχει στο τέλος πιθανότητα K/N, όπου N είναι το (άγνωστο) συνολικό μήκος του ρεύματος.


```python
import random
import sys
import os

def sample(filename, K):
    if K <= 0:         #elegxos na mhn einai arnhtikos arithmos
        print("Σφάλμα: Το K πρέπει να είναι θετικός ακέραιος.", file=sys.stderr)
        return []
    reservoirS = []   #thesimo pinaka toy reservoir
    if len(reservoirS)>K:
        print("σφαλμα δε ισχυει οτι εχουμε Ο(κ)")  #elegxos gia na isxyei oti einai megethos 0(k)

    n = 0  #metritis-antikeimenwn

    if not os.path.exists(filename):   #elegxos an de yparxei to file me ta data
            raise FileNotFoundError(f"Το αρχείο '{filename}' δεν βρέθηκε.")

    with open(filename, 'r', encoding='utf-8') as f:     #ektelesi mono read kai de kanoyme kamia apothikeysi toy arxeioy

            for line in f:      #diatrexei grammi-grammi
                n += 1    #enimwrsi metriti
                clean_line = line.strip()
                if n <= K:
                    reservoirS.append(clean_line)   #gemizei to reservoir me ta prwta k antikeimena toy stream
                else:
                    j = random.randrange(n)      #omioomorfi epilogi tyxaias thesi gia allai apo 0 ews n-1
                    if j < K:
                        reservoirS[j] = clean_line  #antikatastasi toy palioy stoixeioy me to neo sthn j thesi toy reservoir

    return reservoirS   #epistrofi twn antikeimenwn mesa sti dejameni
```

```python

if __name__ == "__main__":    #edw arxizei h main
    filez = "input-A.csv"   #arxeio twn dedomenwn
    Kcount = 0  # arxikopoisi plhthoys k
    while True:
        k_in = input("Δώστε τον επιθυμητό αριθμό δειγμάτων (K): ")  #orismos input gia to arithmo toy K
        Kcount = int(k_in)
        if Kcount > 0:
                break
        else:
                print("Σφάλμα: Ο αριθμός K πρέπει να είναι θετικός ")


    print(f"Αρχείο: {filez}")

    print(f"Ζητούμενο μέγεθος δείγματος (K): {Kcount}")


    randomsample = sample(filez, Kcount)   #kaloyme thn sample me orisma to arxeio kai to arithmo twn K

    if randomsample:
        print(f"\n Το τυχαίο δείγμα των {len(randomsample)} γραμμών είναι:\n")

        for i, line in enumerate(randomsample, start=1):  #kanei ektypwsi twn dedomenwn grammi pros grammi kai me arithmisi gia kathe line
            print(f"[{i:2d}]: {line}")
    else:
        print("\nτο αρχείο ήταν κενό.")
```

Αντιστοίχιση περιγραφής – κώδικα Sample

Για την περιγραφή του κώδικα της def sample, κατά την οποία υλοποιούμε τον αλγόριθμο του Reservoir Sampling, είναι η εξής:

Στις γραμμές 5–8, γίνεται η αρχικοποίηση της δεξαμενής reservoir = [] και του μετρητή n = 0, που υλοποιούν ακριβώς τη θεωρητική δεξαμενή και τον μετρητή του αλγορίθμου.

Στις γραμμές 13–17, μέσα στη for line in f:, αυξάνεται ο μετρητής n κατά 1 για κάθε νέο αντικείμενο (n += 1).
Αν n <= K, το αντικείμενο προστίθεται στη δεξαμενή με reservoir.append(clean_line).
Έτσι, τα Κ πρώτα αντικείμενα του ρεύματος αποθηκεύονται αυτόματα στη δεξαμενή.
Αν το ρεύμα σταματούσε τώρα, κάθε ένα από αυτά τα Κ αντικείμενα θα είχε πιθανότητα επιλογής 1, δηλαδή P=K/n=1

Στις γραμμές 18–21, εφόσον n > K, επιλέγεται ένας τυχαίος ακέραιος j = random.randrange(n), ομοιόμορφα από το διάστημα [0, n−1].
Η συνθήκη if j < K: υλοποιεί ακριβώς την πιθανότητα επιλογής P(επιλογης)=K/n,
αφού υπάρχουν Κ “επιτυχείς” τιμές (0 έως K−1) στις n συνολικές δυνατές τιμές του j.
Άρα, το κάθε νέο αντικείμενο έχει πιθανότητα K/n να επιλεγεί για είσοδο στη δεξαμενή και πιθανότητα 1−K/n να αγνοηθεί

Η γραμμή 21 (reservoir[j] = clean_line) πραγματοποιεί την αντικατάσταση του j-οστού στοιχείου στη δεξαμενή.
Επειδή το j είναι τυχαίος ακέραιος μεταξύ 0 και n−1, και η αντικατάσταση γίνεται μόνο όταν j<K,
κάθε θέση της δεξαμενής έχει ίση πιθανότητα 1/K να αντικατασταθεί από το νέο αντικείμενο.
Έτσι, η αντικατάσταση είναι ομοιόμορφη ως προς τα υπάρχοντα δείγματα.

Η επανάληψη των γραμμών 13–21 εφαρμόζει διαδοχικά τα βήματα 3 και 4 για κάθε νέο αντικείμενο του αρχείου.
Όταν ολοκληρωθεί η ανάγνωση, η εντολή return reservoir στη γραμμή 23 επιστρέφει τη δεξαμενή.
Κατά την ολοκλήρωση, κάθε αντικείμενο από το αρχικό ρεύμα έχει πιθανότητα επιλογής K/N να ανήκει στη δεξαμενή, όπως ακριβώς αποδεικνύεται θεωρητικά στον αλγόριθμο Reservoir Sampling.
